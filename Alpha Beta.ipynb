```python
for x in ticker :
    
    df = yf.download(x,start,end)
    df['Date'] = df.index
    print(x)
    table = pd.DataFrame([[x,
                           df.tail(1).Date[0],
                           df.tail(30).Date[0],
                           df.tail(90).Date[0],
                           df.tail(180).Date[0],
                           (df.tail(1).values[:1,3][0] - df.tail(30).head(1).values[:1,3][0])/df.tail(30).head(1).values[:1,3][0],
                           (df.tail(1).values[:1,3][0] - df.tail(90).head(1).values[:1,3][0])/df.tail(90).head(1).values[:1,3][0],
                           (df.tail(1).values[:1,3][0] - df.tail(180).head(1).values[:1,3][0])/df.tail(180).head(1).values[:1,3][0]
                          ]],
                     columns =['Ticker','Last_Dt','Date_30','Date_90','Date_180','Returns_30_Days','Returns_90_Days','Returns_180_Days'])
    fin = fin.append(table)
```

```python
nifty = yf.download('^NSEI',start,end)
```

```python
nifty.tail(5)
```

```python
##30 working days nifty returns
nifty_ret_30 = (nifty.tail(1).values[:1,3][0] - nifty.tail(30).head(1).values[:1,3][0])/nifty.tail(30).head(1).values[:1,3][0]
nifty_ret_90 = (nifty.tail(1).values[:1,3][0] - nifty.tail(90).head(1).values[:1,3][0])/nifty.tail(90).head(1).values[:1,3][0]
nifty_ret_180 = (nifty.tail(1).values[:1,3][0] - nifty.tail(180).head(1).values[:1,3][0])/nifty.tail(180).head(1).values[:1,3][0]
```

```python
fin['alpha_30'] = fin.Returns_30_Days/nifty_ret_30
fin['alpha_90'] = fin.Returns_90_Days/nifty_ret_90
fin['alpha_180'] = fin.Returns_180_Days/nifty_ret_180
```

```python
fin.to_csv('alpha_v2.csv')
```
