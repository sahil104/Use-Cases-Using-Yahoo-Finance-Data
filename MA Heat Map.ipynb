```python
for x in ticker :
    
    df = yf.download(x,start= start_d,end = end_d, interval = '1d' )
    df['Date'] = df.index
    df['Ticker'] = x
    table = df
    table['EMA_8_d'] = df['Close'].ewm(span=8).mean()
    table['EMA_21_d'] = df['Close'].ewm(span=21).mean()
    table['EMA_60_d'] = df['Close'].ewm(span=69).mean()
    table['EMA_100_d'] = df['Close'].ewm(span=100).mean()
    table['EMA_200_d'] = df['Close'].ewm(span=200).mean()
    fin_d = fin_d.append(table)
    
fin_d['P_minus_MA8_d'] = (fin_d['Close'] - fin_d['EMA_8_d'])/fin_d['Close']
fin_d['P_minus_MA21_d'] = (fin_d['Close'] - fin_d['EMA_21_d'])/fin_d['Close']
fin_d['P_minus_MA60_d'] = (fin_d['Close'] - fin_d['EMA_60_d'])/fin_d['Close']
fin_d['P_minus_MA100_d'] = (fin_d['Close'] - fin_d['EMA_100_d'])/fin_d['Close']
fin_d['P_minus_MA200_d'] = (fin_d['Close'] - fin_d['EMA_200_d'])/fin_d['Close']    
```

```python
for x in ticker :
    
    df = yf.download(x,start= "2021-01-01",end = end_h, interval = '1h' )
    df['Date'] = df.index
    df['Ticker'] = x
    table = df
    table['EMA_8_h'] = df['Close'].ewm(span=8).mean()
    table['EMA_21_h'] = df['Close'].ewm(span=21).mean()
    table['EMA_60_h'] = df['Close'].ewm(span=69).mean()
    table['EMA_100_h'] = df['Close'].ewm(span=100).mean()
    table['EMA_200_h'] = df['Close'].ewm(span=200).mean()
    fin_h = fin_h.append(table)

fin_h['P_minus_hA8_h'] = (fin_h['Close'] - fin_h['EMA_8_h'])/fin_h['Close']
fin_h['P_minus_hA21_h'] = (fin_h['Close'] - fin_h['EMA_21_h'])/fin_h['Close']
fin_h['P_minus_hA60_h'] = (fin_h['Close'] - fin_h['EMA_60_h'])/fin_h['Close']
fin_h['P_minus_hA100_h'] = (fin_h['Close'] - fin_h['EMA_100_h'])/fin_h['Close']
fin_h['P_minus_hA200_h'] = (fin_h['Close'] - fin_h['EMA_200_h'])/fin_h['Close']    
```

```python
for x in ticker :
    
    df = yf.download(x,start= start_m,end = end_m, interval = '5m' )
    df['Date'] = df.index
    df['Ticker'] = x
    table = df
    table['EMA_8_m'] = df['Close'].ewm(span=8).mean()
    table['EMA_21_m'] = df['Close'].ewm(span=21).mean()
    table['EMA_60_m'] = df['Close'].ewm(span=69).mean()
    table['EMA_100_m'] = df['Close'].ewm(span=100).mean()
    table['EMA_200_m'] = df['Close'].ewm(span=200).mean()
    fin_m = fin_m.append(table)
    
fin_m['P_minus_MA8_m'] = (fin_m['Close'] - fin_m['EMA_8_m'])/fin_m['Close']
fin_m['P_minus_MA21_m'] = (fin_m['Close'] - fin_m['EMA_21_m'])/fin_m['Close']
fin_m['P_minus_MA60_m'] = (fin_m['Close'] - fin_m['EMA_60_m'])/fin_m['Close']
fin_m['P_minus_MA100_m'] = (fin_m['Close'] - fin_m['EMA_100_m'])/fin_m['Close']
fin_m['P_minus_MA200_m'] = (fin_m['Close'] - fin_m['EMA_200_m'])/fin_m['Close']
```

```python
for x in ticker:

    dly = fin_d[fin_d['Ticker'] == x].tail(1).iloc[:,[0,13,14,15,16,17]]
    dly['Pd'] = 'Daily'
    hrly = fin_h[fin_h['Ticker'] == x].tail(1).iloc[:,[0,13,14,15,16,17]]
    hrly['Pd'] = 'Hourly'
    mint = fin_m[fin_m['Ticker'] == x].tail(1).iloc[:,[0,13,14,15,16,17]]
    mint['Pd'] = 'Five_Minute'

    dly.columns = ['Ticker','P_by_EMA_8','P_by_EMA_21','P_by_EMA_60','P_by_EMA_100','P_by_EMA_200','Pd']
    hrly.columns = ['Ticker','P_by_EMA_8','P_by_EMA_21','P_by_EMA_60','P_by_EMA_100','P_by_EMA_200','Pd']
    mint.columns = ['Ticker','P_by_EMA_8','P_by_EMA_21','P_by_EMA_60','P_by_EMA_100','P_by_EMA_200','Pd']



    ultimate = ultimate.append(dly)
    ultimate = ultimate.append(hrly)
    ultimate = ultimate.append(mint)
```
