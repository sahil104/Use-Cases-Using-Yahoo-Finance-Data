```python
import datetime as dt
import matplotlib.pyplot as plt
from matplotlib import style
import datetime as dt
import numpy as np
import yfinance as yf
import pandas as pd
```

```python
from datetime import datetime, timedelta
tomorrow = datetime.today() + timedelta(1)
tomorrow.strftime('%Y-%m-%d')
```

```python
start= "2015-02-01"
end = tomorrow.strftime('%Y-%m-%d')
```

```python

fin = pd.DataFrame(columns =['Ticker', 'Min', 'Max'])
```

```python
##data pull for tickers from yahoo finance##
ticker = ['HDFC.NS', 'HDFCLIFE.NS', 'IEX.NS', 'IRFC.NS',  'EMBASSY.BO', 'ZOMATO.NS', 'NAUKRI.NS', 'MASTEK.NS', 'MINDTREE.NS', 'TANLA.NS', 'TCS.NS', 'LAURUSLABS.NS', 'NH.NS', 'ADANIGREEN.NS', 'ADANIPORTS.NS', 'BHARTIARTL.NS', 'HFCL.NS', 'ROUTE.NS']
```

```python
df_nifty = yf.download('^NSEI',start= "2018-02-01",end = tomorrow.strftime('%Y-%m-%d'), interval = '1h' )
```

```python
for x in ticker :
    
    df = yf.download(x,start,end, interval = '1d' )
    df['Date'] = df.index
    print(x)
    table = pd.DataFrame([[x,
                       df.Date[len(df)-1],
                       df.Date[0],
                       df.Close.min(),
                       df.Close.max(),
                       df.tail(1).Close[0], 
                       df.tail(10).Close.mean(),
                       df.tail(50).Close.mean(),
                       df.tail(100).Close.mean(),
                       df.tail(200).Close.mean(),
                       (df.tail(1).values[:1,3][0] -  df.tail(30).values[:1,3][0])/df.tail(30).values[:1,3][0],
                       (df.tail(1).values[:1,3][0] -  df.tail(100).values[:1,3][0])/df.tail(100).values[:1,3][0],
                       (df.tail(1).values[:1,3][0] -  df.tail(365).values[:1,3][0])/df.tail(365).values[:1,3][0]]],
                     columns =['Ticker','Last_Dt','First_Dt', 'Min', 'Max','Last Close','MA_10','MA_50','MA_100','MA_200','Delta_30_D','Delta_100_D','Delta_365_D'])
    fin = fin.append(table)
```

```python
fin
```

```python
fin['Buy_Price'] = [2941.1, 579.14, 268.81, 26, 340.9, 76, 5024.44, 2513.38, 2618.29, 1801.09, 3867.98, 525.17, 665.74, 1438.8, 719.42, 537.81, 97.48, 1810.09]
fin['Buy_Qty'] = [10, 50, 110, 575,  160, 195, 6, 10, 10, 13, 8, 68, 45, 35, 42, 40, 300, 18]
fin['Buy_Price'] = pd.to_numeric(fin.Buy_Price)
fin['Buy_Qty'] = pd.to_numeric(fin.Buy_Qty)
fin['Total_Cost'] = fin['Buy_Price']*fin['Buy_Qty']
fin['Total_Cost'] = pd.to_numeric(fin.Total_Cost)
fin['Total_Current'] = fin['Last Close']*fin['Buy_Qty']
fin['Profit'] = (fin['Last Close'] - fin['Buy_Price'])*fin['Buy_Qty']
```

```python
fin['Allocation_Perc_Cost'] = fin['Total_Cost']/sum(fin['Total_Cost'])
fin['Allocation_Perc_Curr'] = fin['Total_Current']/sum(fin['Total_Current'])
```

```python
fin = fin.sort_values(by = ['Allocation_Perc_Cost'], ascending = False)
```

```python
fin.to_csv('Portfolio_v2.csv')
```

```python
fin['P_minus_MA10'] = (fin['Last Close'] - fin['MA_10'])/fin['Last Close']
fin['P_minus_MA50'] = (fin['Last Close'] - fin['MA_50'])/fin['Last Close']
fin['P_minus_MA100'] = (fin['Last Close'] - fin['MA_100'])/fin['Last Close']
fin['P_minus_MA200'] = (fin['Last Close'] - fin['MA_200'])/fin['Last Close']
```
