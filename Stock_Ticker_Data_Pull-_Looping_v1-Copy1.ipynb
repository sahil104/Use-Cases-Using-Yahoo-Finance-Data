```python
for x in ticker :
    
    df = web.DataReader(x,'yahoo',start,end)
    df['Date'] = df.index
    print(x)
    table = pd.DataFrame([[x,
                       df.Date[len(df)-1],
                       df.Date[0],
                       df.Close.min(),
                       df.Close.max(),
                       df.tail(1).Close[0], 
                       df.tail(10).Close.mean(),
                       df.tail(25).Close.mean(),
                       df.tail(60).Close.mean(),
                       df.tail(100).Close.mean(),
                       (df.tail(1).values[:1,3][0] -  df.tail(30).values[:1,3][0])/df.tail(30).values[:1,3][0],
                       (df.tail(1).values[:1,3][0] -  df.tail(100).values[:1,3][0])/df.tail(100).values[:1,3][0],
                       (df.tail(1).values[:1,3][0] -  df.tail(365).values[:1,3][0])/df.tail(365).values[:1,3][0]]],
                     columns =['Ticker','Last_Dt','First_Dt', 'Min', 'Max','Last Close','MA_10','MA_25','MA_60','MA_100','Delta_30_D','Delta_100_D','Delta_365_D'])
    fin = fin.append(table)
```
